### Импорт необходимых библиотек

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Загрузка данных

```python
df = pd.read_csv("/content/Admission_Predict.csv")
df.head()
```

### Корреляционный анализ

```python
df.info()
```

Как видно, пропусков в данных нет

```python
df.describe()
```

Удалим Serial No. так как это бесполезный признак(Это просто номер, который ничего не значит для анализа данных)

```python
df = df.drop(columns=['Serial No.'])
df.head()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

plt.title('Матрица корреляции')
plt.xlabel('Переменные')
plt.ylabel('Переменные')

plt.show()
```

Как видно из матрицы корреляции, каждый признак имеет значение для предсказания таргета.

Первое, чтобы я сделал в таком случае - это построил линейную регрессию (так как у нас задача регресси)

Если бы качество модели было бы неудовлетворительным, то попробовал бы дерево решений, ансабль деревьев - решающий лес для задачи регрессии
